# v22 — Comparable Sales + Floor Premium Features

Adds temporal comparable-sales features to the pipeline:
- storey_pct — relative floor position within the building
- comp_6m_count — # transactions, same (town, flat_model), past 6 months
- comp_6m_median — median resale price of those comps
- comp_12m_median — same over 12-month window (smoother for rare models)
- comp_floor_adj_median — comp_6m_median adjusted for floor level difference (+$4K/floor)

These features encode buyer/seller appraisal logic: compare against recent, same-model,
same-area transactions, adjusting for floor premium.
